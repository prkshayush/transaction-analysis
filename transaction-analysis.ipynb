```python
import sqlalchemy as sqa
engine = sqa.create_engine('mssql://MSI-AP/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn = engine.connect()
```

```python
df.to_sql('df_transactions', con=conn, index=False, if_exists = 'replace')
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
query = ''' 
SELECT category, sum(transaction_amount) as sales
FROM df_transactions
GROUP BY category
ORDER BY sales desc
'''
df_query_1 = pd.read_sql_query(query, con=conn)
```

```python
df_query_1
```

```python
df_query_1['category']
```

```python
y = []
for category in df_query_1['category']:
    y.append(category)
print(y)
```

```python
pie_1 = []
for sales in df_query_1['sales']:
    pie_1.append(sales)
print(pie_1)
```

```python
data = np.array(pie_1)
plt.pie(data, labels=y, autopct='%1.1f%%')
plt.show()
```

```python
query_2 = ''' 
SELECT gender, sum(transaction_amount) as sales
FROM df_transactions
GROUP BY gender
ORDER BY sales desc
'''
df_query_2 = pd.read_sql_query(query_2, con=conn)
```

```python
y = []
for category in df_query_2['gender']:
    y.append(category)
print(y)
```

```python
pie_2 = []
for sales in df_query_2['sales']:
    pie_2.append(sales)
print(pie_2)
```

```python
data = np.array(pie_2)
plt.pie(data, labels=y, autopct='%1.1f%%')
plt.show()
```

```python
query_3 = ''' 
SELECT distinct month(date) as month, sum(transaction_amount) as sales
FROM df_transactions
GROUP BY month(date)
ORDER BY month, sales asc
'''
df_query_3 = pd.read_sql_query(query_3, con=conn)
```

```python
df_query_3
```

```python
y = []
for sales in df_query_3['sales']:
    y.append(sales)

x = []
for month in df_query_3['month']:
    x.append(month)

xpoints = np.array(x)
ypoints = np.array(y)

plt.xlabel("Months")
plt.ylabel("Total sales")

plt.plot(xpoints, ypoints, marker = 'o', ms = 5, mfc = 'r')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```
